```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')

df.head(10).plot.barh('Geography','Age')  #creating a bar pyplot



```

```
data_pie=df.groupby('Geography').NumOfProducts.count()
plt.pie(np.array(data_pie),labels=('france','spain','Germany'),explode=(0.05,0.05,0.05),shadow=True,autopct='%d%%')         # Creating a pie chart


```

```

df.tail(10).boxplot(column="EstimatedSalary",by='Geography')    # creating a box plot

```

```
df.head(50).CreditScore.plot(kind='kde')       # creating a kernal density graph
```

```
# scatter plot between geography and age 
plt.scatter(df['Geography'], df['Age']) 
plt.show() 
  
```
